```python
import pandas as pd

adult_data = pd.read_csv("adult_data.csv", header=None, names=["age", "workclass", "fnlwgt", "education", "education_num", "marital_status", "occupation", "relationship", "race", "sex", "capital_gain", "capital_loss", "hours_per_week", "native_country", "class"], index_col=False)
```

```python
adult_data.tail()
```

```python
eq_width_bin = pd.cut(adult_data["age"], bins=4)
```

```python
adult_data["eq_width_bin"] = eq_width_bin
```

```python
adult_data.head()
```

```python
eq_deepth_bin = pd.qcut(adult_data["age"], q=4)
```

```python
adult_data["eq_deepth_bin"] = eq_deepth_bin
```
